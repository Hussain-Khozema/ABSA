```python
# We get started by importing spacy
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def identify_aspects(sentence):
  doc = nlp(sentence)
  descriptive_term = ''
  target = ''
  for token in doc:
    if token.dep_ == 'nsubj' and token.pos_ == 'NOUN':
      target = token.text
    if token.pos_ == 'ADJ':
      prepend = ''
      for child in token.children:
        if child.pos_ != 'ADV':
          continue
        prepend += child.text + ' '
      descriptive_term = prepend + token.text
  return {'aspect': target, 'description': descriptive_term}
```

```python
identify_aspects("This has a good battery life")
```

### Lets try on Laptop Reviews

```python
laptop_reviews_path = 'sameval2014/Train_Data_v2.0/Laptop_Train_v2.xml'
```

```python
import xml.etree.ElementTree as ET
from string import punctuation

tree = ET.parse(laptop_reviews_path)

root = tree.getroot()

print(root.tag)
```

```python
sentences = {}
for sentence in root:
  text = sentence.find("text").text
  aspects = []
  try:
    for child in sentence.find('aspectTerms'):
      aspects.append(child.attrib['term'])
  except:
    continue
  sentences[text] = aspects

print(len(sentences))
```

```python
count = 10

for key,value in sentences.items():
  print(identify_aspects(key))
  break
```
